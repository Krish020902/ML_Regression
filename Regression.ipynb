```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def generate_sample_data():
    # Generating sample data
    np.random.seed(0)
    x_train = np.random.rand(50, 1) * 10
    y_train = 3 * x_train + np.random.randn(50, 1)
    return x_train, y_train

def display_data(x_train, y_train):
    df = pd.DataFrame({'Feature': x_train.flatten(), 'Output': y_train.flatten()})
    print(df)

def fit_linear_regression(x_train, y_train):
    # Add bias term to x_train
    x_train = np.concatenate((np.ones((x_train.shape[0], 1)), x_train), axis=1)
    
    # Calculate the coefficients using closed-form solution
    coefficients = np.linalg.inv(x_train.T.dot(x_train)).dot(x_train.T).dot(y_train)
    
    return coefficients

def predict(x_test, coefficients):
    # Add bias term to x_test
    x_test = np.concatenate((np.ones((x_test.shape[0], 1)), x_test), axis=1)
    
    # Make predictions
    predictions = x_test.dot(coefficients)
    
    return predictions

def plot_regression_line(x_train, y_train, x_test, y_test, coefficients):
    # Plot training data
    plt.scatter(x_train, y_train, label='Training Data')
    
    # Plot test data
    plt.scatter(x_test, y_test, label='Test Data')
    
    # Plot regression line
    plt.plot(x_train, x_train * coefficients[1] + coefficients[0], color='red', label='Regression Line')
    
    plt.xlabel('Feature')
    plt.ylabel('Output')
    plt.title('Linear Regression')
    plt.legend()
    plt.show()

# Generate sample data
x_train, y_train = generate_sample_data()

# Display the sample data
display_data(x_train, y_train)

# Fit linear regression model
coefficients = fit_linear_regression(x_train, y_train)

# Test data
x_test = np.array([[1.5], [5.0], [8.2]])  # You can modify the test data as per your needs

# Perform predictions and plot results
predictions = predict(x_test, coefficients)
plot_regression_line(x_train, y_train, x_test, predictions, coefficients)
```
